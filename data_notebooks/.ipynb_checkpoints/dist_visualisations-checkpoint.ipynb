```python
import numpy as np

class simple_generation:

    def __init__(self, mean, std, no_samples):
        
        self.mean = mean
        self.std = std
        self.no_samples = no_samples
        self.init_dist = np.random.normal(mean, std, no_samples)
        
    def init_dist(self):

        return self.init_dist
    
    def quad_fit(self):
        
        c = np.log(1/(np.sqrt(2*np.pi)*self.std))
        self.c = c
        quad_function = lambda x: c - ((x-self.mean)**2)/(2*self.std**2)
        quad_dist = quad_function(self.init_dist)

        self.quad_function = quad_function
        
        return quad_dist

    def linear_fit(self):
        
        lin_function = np.sqrt(-self.quad_fit() + self.c)
        return lin_function
    
    def c_return(self):
        return self.c
```

```python
import matplotlib.pyplot as plt
```

```python
a = simple_generation(0, 1,10000)
plt.hist(a.init_dist)
```

```python
plt.hist(a.quad_fit())
```

```python
plt.hist(a.linear_fit())
```

```python
x = a.init_dist
y = a.quad_fit()
plt.plot(x, y, ".")
```

```python
b = a.linear_fit()
c = a.c_return()

d = b**2

plt.plot(x, b, ".")
plt.figure()
plt.plot(x,-d,'.')
```
